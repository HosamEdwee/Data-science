#Import turicreate and Load the house prices dataset

```python
import turicreate
sales = turicreate.SFrame('drive/MyDrive/data/home_data.sframe')
```

#Explore data

```python
sales #print the head of the dataframe
```

```python
sales.show()  #visualize data
```

```python
#We should do correlation test first, but for simplicity, we will assume our variables are associated with the response.
turicreate.visualization.scatter(x=sales['sqft_living'][0:100], y = sales['price'][0:100]) #I chose only the first 100 element to save resources
```

#Split the data into training, test set

```python
training_set, test_set = sales.random_split(.8,seed=0) #I will use seed 0 to fix the randomness factor
```

Train linear regression model

```python
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

Evaluate the quality of the model

```python
print (test_set['price'].mean())
```

```python
print (sqft_model.evaluate(test_set))
```

# Get the model coefficients





```python
sqft_model.coefficients
```

# Plot the model against test data

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_set['sqft_living'],test_set['price'],'.',
        test_set['sqft_living'],sqft_model.predict(test_set),'-')
```

#explore other features

# Build a model with these additional features

```python
my_features_model = turicreate.linear_regression.create(training_set,target='price',features=my_features)
```

# Compare the two models

```python
print (sqft_model.evaluate(test_set))
print (my_features_model.evaluate(test_set))
```
